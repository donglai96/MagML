The step1 is a basic quick method to visualize the omni data, which is the driver we will use in the following tutorials.
The OMNI data (https://omniweb.gsfc.nasa.gov/ow_min.html) include Solar wind magnetic field and plasma data at Earth's Bow Shock Nose (BSN), also geomagnetic activity indices and 5-min energetic proton fluxes, which is a perfect driving parameter and database for magnetosphere studies.
The code here is just a warm-up, and simple wrapper of pyspedas (https://github.com/spedas/pyspedas) which can fill the gap of data.

```python
import magml.data.omni as omni

data_total = omni.load.InputVar(name = ['AE_INDEX','AL_INDEX','SYM_H','flow_speed','Pressure','BZ_GSM'],
                                                   start_time= "2012-03-01",
                                                   end_time= "2012-04-01",
                                                   no_update = False,data_type='5min',gap_max=180)
```

```python
AE = data_total.get_data('AE_INDEX')
SYM_H = data_total.get_data('SYM_H')
```

```python
import matplotlib.pyplot as plt

fig,axs = plt.subplots(2,1,figsize=(10,10))
axs[0].plot(AE)
axs[1].plot(SYM_H)
plt.show()
```
